```python
df_reviews['tokened'] = df_reviews.review.apply(word_tokenize)
```

```python
df_reviews.head()
```

```python
def to_lowercase(words):
    minusculas = [i.lower() for i in words]
    
    return minusculas
```

```python
df_reviews['normal'] = df_reviews.tokened.apply(to_lowercase)
df_reviews.head()
```

```python
def removePunc(wordList):
    palabras_list=[]
#     puncList = [".",";",":","!","?","/","\\",",",")","(","\""]
    puncList = [["."],[";"],[":"],["!"],["?"],["/"],[","],[")"],["("],["..."],[".."],[", "],[",,"]]
    lista = [i.strip().split(',') for i in wordList]
    
    for element in lista:
        palabras = element 
        
        if palabras in puncList:
            pass
        else:
            palabras_list.append(palabras)
            
            
                                            
    return palabras_list
```

```python
df_reviews['new_normal'] = df_reviews.normal.apply(removePunc)
```

```python
df_reviews.drop(['new_normal'], axis=1)
```

```python
def remove_stopwords(words):
    stop = [word for word in words if word not in stopwords.words('spanish')]
    return stop
```

```python
df_reviews['nonstop'] = df_reviews.normal.apply(remove_stopwords)
df_reviews.head()
```

```python
df_reviews['nopoint'] = df_reviews.nonstop.apply(removePunc)
df_reviews.head()
```
